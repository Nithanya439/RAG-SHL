```python
from langchain_community.document_loaders import WebBaseLoader

url = "https://www.shl.com/products/product-catalog/"
loader = WebBaseLoader(url)
docs = loader.load()

print(f"Total docs loaded: {len(docs)}")
print(docs[0].page_content[:1000])
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
splits = splitter.split_documents(docs)

print(f"Total chunks: {len(splits)}")
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
from langchain_community.vectorstores import Chroma

vectorstore = Chroma.from_documents(splits, embedding=embedding_model, persist_directory="shl_chroma_db")
```

```python
retriever = vectorstore.as_retriever(search_kwargs={"k": 5})
```

```python
from langchain_community.llms import HuggingFacePipeline
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline

model_name = "google/flan-t5-base"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

pipe = pipeline("text2text-generation", model=model, tokenizer=tokenizer, max_length=512)

llm = HuggingFacePipeline(pipeline=pipe)
```

```python
from langchain.chains import RetrievalQA

rag_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever,
    return_source_documents=True
)
```

```python
query = "What types of assessment solutions does SHL provide?"
result = rag_chain.invoke(query)

print("🔹 Answer:\n", result["result"])


```
